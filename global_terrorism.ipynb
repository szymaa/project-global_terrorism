```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data_terrorism=pd.read_csv('globalterrorismdb_0617dist.csv',encoding = 'ISO-8859-1', usecols=[1,7,8,9,10,28,29,34,35,36])

#1 - Countries with the highest number of terrorist attacks (1970-2016)

attack_counts = data_terrorism['country_txt'].value_counts().nlargest(10)
plt.figure(facecolor='#F7F6F6', figsize=(15,6))
x = range(len(attack_counts))
y = attack_counts
plt.bar(x, y, color='#6690AD')
plt.title("Countries with the highest number of terrorist attacks (1970-2016)",fontsize=15)
plt.ylabel("Number of attacks")
plt.xticks(range(len(attack_counts)), attack_counts.keys())
plt.show()
print(attack_counts)
```

```python
#2 - The total number of terrorist attacks worldwide in 1970-2016

plt.figure(facecolor='#F7F6F6', figsize=(15,6))

df = pd.DataFrame(data_terrorism, columns=['iyear'])
plt.title("The total number of terrorist attacks worldwide in 1970-2016",fontsize=15)
df['iyear'].groupby([df.iyear]).count().plot.bar(color='#6690AD')
plt.ylabel("Number of attacks")
plt.xlabel("Year")

plt.show()
```

```python
#3 - The total number of terrorist attacks per country in South America (1970-2016)

attack_counts_SA = data_terrorism.loc[data_terrorism['region_txt']=='South America']
country_attacks_SA = attack_counts_SA['country_txt'].value_counts()
plt.figure(facecolor='#F7F6F6', figsize=(15,6))
x = range(len(country_attacks_SA))
y = country_attacks_SA
plt.bar(x, y, color='#6690AD')
plt.title("The total number of terrorist attacks per country in South America (1970-2016)",fontsize=15)
plt.ylabel("Number of attacks")
plt.xticks(range(len(country_attacks_SA)), country_attacks_SA.keys(), rotation=20)
plt.show()
print(country_attacks_SA)
```

```python
#4 - The most common type of terrorist attack in South America (1970-2016)

attack_counts_SA = data_terrorism.loc[data_terrorism['region_txt']=='South America']
type_of_attack = attack_counts_SA.groupby(['attacktype1'])
type_of_attack = attack_counts_SA['attacktype1_txt'].value_counts()
plt.figure(facecolor='#F7F6F6', figsize=(15,6))
x = range(len(type_of_attack))
y = type_of_attack
plt.bar(x, y, color='#6690AD')
plt.title("The most common type of terrorist attack in South America (1970-2016)",fontsize=15)
plt.ylabel("Number of attacks")
plt.xticks(range(len(type_of_attack)), type_of_attack.keys(), rotation=70)
plt.show()
print(type_of_attack)
```

```python
#5 Information about target/victim terrorist attack in Poland (1970-2016)
plt.figure(facecolor='#F7F6F6', figsize=(15,6))
attack_counts_Poland = data_terrorism.loc[data_terrorism['country_txt']=='Poland']
target_attack = attack_counts_Poland['targtype1_txt'].value_counts()
df = pd.DataFrame(attack_counts_Poland, columns=['country_txt','targtype1_txt'])
plt.title("Information about target/victim of terrorist attack in Poland (1970-2016)",fontsize=15)

df.groupby('targtype1_txt').size().plot(kind='pie',autopct='%1.2f%%',shadow=None, startangle=9).axis('equal')
plt.show()
print(target_attack)
```
